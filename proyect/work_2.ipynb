## 0. Introduction

In today’s digital age, the increasing prevalence of problematic internet use among children and adolescents poses significant challenges to mental health and well-being. Conditions such as depression and anxiety are often linked to excessive technology usage, highlighting the importance of early detection and intervention.

This notebook is part of a competition designed to address this critical issue by developing a predictive model that analyzes children's physical activity data. The goal is to identify early indicators of problematic internet and technology use, enabling timely interventions to foster healthier digital habits and mitigate mental health risks.

The data for this competition comes from the Healthy Brain Network, a groundbreaking mental health study conducted in New York City. This initiative, led by the Child Mind Institute, collaborates with families, community leaders, and supporters to advance our understanding of the developing brain. Financial support for this initiative has been provided by the California Department of Health Care Services (DHCS) as part of the Children and Youth Behavioral Health Initiative (CYBHI), with additional contributions from the Kaggle team.
